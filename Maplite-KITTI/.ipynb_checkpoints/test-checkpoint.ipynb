```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import scipy.signal as sig
import time
```

```python
imu = pd.read_csv("./../Datasets/HVL-multi-loops/HVL-multi-loops/imu.csv")
enc = pd.read_csv("./../Datasets/HVL-multi-loops/HVL-multi-loops/wheel_control.csv")
imu = np.array(imu)
enc = np.array(enc)
enc = enc[1884:]
imu = imu[31:]

n=imu.shape[0]
m=enc.shape[0]

a_x=imu[:,2]
a_y=imu[:,3]
a_z=imu[:,4]

bias_x=np.mean(a_x[:9500])
bias_y=np.mean(a_y[:9500])

a_x=a_x-bias_x
a_y=a_y-bias_y

g=np.mean(a_z)
print("g = %.9f"%(g))

medax=sig.medfilt(a_x,1001)
meday=sig.medfilt(a_y,1001)
medaz=sig.medfilt(a_z,1001)
lpf_a=sig.firwin(801, 0.01, window='hamming')

testa_x=sig.convolve(medax,lpf_a,mode='same')
testa_y=sig.convolve(meday,lpf_a,mode='same')
testa_z=sig.convolve(medaz,lpf_a,mode='same')

theta=np.zeros(n)
phi=np.zeros(n)

for i in range(n):
    if i==0:
        continue
    theta[i]=theta[i-1]+(0.01*enc[i][3])
    phi[i]=phi[i-1]+(0.01*imu[i][7])
```

```python
fig,(ax,ay,az)=plt.subplots(3,1,sharex=True,figsize=(30,21))
fig.suptitle("Raw data",fontweight="bold",fontsize=40)
ax.scatter(range(n),a_x,s=0.1)
ax.scatter(range(n),testa_x,s=0.1)
ax.set_ylabel("acc_x",fontsize=25)
ay.scatter(range(n),a_y,s=0.1)
ay.scatter(range(n),testa_y,s=0.1)
ay.set_ylabel("acc_y",fontsize=25)
az.scatter(range(n),a_z,s=0.1)
az.scatter(range(n),testa_z,s=0.1)
az.set_ylabel("acc_z",fontsize=25)
az.set_xlabel("iter",fontsize=25)
```

```python
plt.figure(figsize=(30,10))
plt.suptitle("Orientation using Gyro",fontsize=30)
plt.scatter(range(n),phi,s=0.1)
plt.xlabel("iter",fontsize=25)
plt.ylabel("Orientation",fontsize=25)
```

```python
plt.figure(figsize=(30,10))
plt.suptitle("Orientation using wheel encoders",fontsize=30)
plt.scatter(range(n),theta,s=0.1)
plt.xlabel("iter",fontsize=25)
plt.ylabel("Orientation",fontsize=25)
```

```python
acc=[]

for i in range(n):
    if i < 9500:
        acc.append(0)
        continue
    temp=np.array([a_x[i],a_y[i],a_z[i]])
    curr_acc=np.linalg.norm(temp)
    curr_acc=curr_acc**2
    curr_acc=abs(curr_acc-(g**2))
    curr_acc=np.sqrt(curr_acc)
    acc.append(curr_acc)

acc=np.array(acc)
```

```python
med=sig.medfilt(acc,31)
lpf_acc=sig.firwin(801, 0.01, window='hamming')

acc=sig.convolve(med,lpf_acc,mode='same')
```

```python
plt.figure(figsize=(30,10))
plt.plot(range(n),acc)
plt.xlabel("iter",fontsize=25)
plt.ylabel("Acc magnitude",fontsize=25)
```

```python
acc_x=[]
acc_y=[]

for i in range(n):
    acc_x.append(acc[i]*np.cos(theta[i]))
    acc_y.append(acc[i]*np.sin(theta[i]))

acc_x=np.array(acc_x)
acc_y=np.array(acc_y)
```

```python
fig1,(ax1,ay1)=plt.subplots(2,1,sharex=True,figsize=(30,15))
fig1.suptitle("Acc after discarding g",fontweight="bold",fontsize=30)
ax1.plot(range(n),acc_x)
ax1.set_ylabel("acc_x",fontsize=25)
ay1.plot(range(n),acc_y)
ay1.set_ylabel("acc_y",fontsize=25)
ay1.set_xlabel("iter",fontsize=25)
```

```python
med_ax=sig.medfilt(acc_x,3001)
med_ay=sig.medfilt(acc_y,3001)
```

```python
fig2,(ax2,ay2)=plt.subplots(2,1,sharex=True,figsize=(30,15))
fig2.suptitle("Median Acc",fontweight="bold",fontsize=30)
ax2.plot(range(n),acc_x)
ax2.plot(range(n),med_ax)
ax2.set_ylabel("acc_x",fontsize=25)
ay2.plot(range(n),acc_y)
ay2.plot(range(n),med_ay)
ay2.set_ylabel("acc_y",fontsize=25)
ay2.set_xlabel("iter",fontsize=25)
```

```python
lpf=sig.firwin(801, 0.01, window='hamming')

proc_ax=sig.convolve(med_ax,lpf,mode='same')
proc_ay=sig.convolve(med_ay,lpf,mode='same')
```

```python
fig3,(ax3,ay3)=plt.subplots(2,1,sharex=True,figsize=(30,15))
fig3.suptitle("Processed FIR filter Acc",fontweight="bold",fontsize=30)
ax3.plot(range(n),acc_x)
ax3.plot(range(n),proc_ax)
ax3.set_ylabel("acc_x",fontsize=25)
ay3.plot(range(n),acc_y)
ay3.plot(range(n),proc_ay)
ay3.set_ylabel("acc_y",fontsize=25)
ay3.set_xlabel("iter",fontsize=25)
```

```python
#Ignore this cell for now
b,a=sig.butter(8,0.05,btype='lowpass')

test_ax=sig.filtfilt(b,a,acc_x)
test_ay=sig.filtfilt(b,a,acc_y)
```

```python
fig4,(ax4,ay4)=plt.subplots(2,1,sharex=True,figsize=(30,15))
fig4.suptitle("Test Butterworth filter Acc",fontweight="bold",fontsize=30)
ax4.plot(range(n),acc_x)
ax4.plot(range(n),test_ax)
ax4.set_ylabel("acc_x",fontsize=25)
ay4.plot(range(n),acc_y)
ay4.plot(range(n),test_ay)
ay4.set_ylabel("acc_y",fontsize=25)
ay4.set_xlabel("iter",fontsize=25)
```

```python
vel=[]
temp_vel=[]
k=[]
vel.append(np.array([0,0,0]))

state=np.array([0,0,0])
thet=0

z=np.array([0,0,0])

P=np.eye(3)

sigma_u=np.eye(3)
sigma_u=sigma_u

R=np.eye(3)
R=R*0.1
```

```python
for i in range(n):
    u=np.array([testa_x[i],testa_y[i],imu[i][7]])
    
    z[0]=enc[i][2]*np.cos(thet)
    z[1]=enc[i][2]*np.sin(thet)
    z[2]=enc[i][3]
    
    G=np.eye(3)
    G=G*0.01
    G[2][2]=1
    
    temp_x=np.zeros(3)
    temp_x[0]=state[0]+(0.01*u[0])
    temp_x[1]=state[1]+(0.01*u[1])
    temp_x[2]=u[2]
    
    temp_vel.append(z)
    
    F=np.eye(3)
    F[2][2]=0
    
    Q=np.dot(G,sigma_u)
    Q=np.dot(Q,np.transpose(G))
    
    temp_p=np.dot(F,P)
    temp_p=np.dot(temp_p,np.transpose(F))
    temp_p=temp_p+Q
    
    H=np.eye(3)
    temp_z=np.dot(H,np.transpose(temp_x))
    
    K=np.dot(temp_p,np.transpose(H))
    temp=np.dot(H,temp_p)
    temp=np.dot(temp,np.transpose(H))
    temp=np.linalg.pinv(temp+R)
    K=np.dot(K,temp)
    k.append(K)
    
    P=np.dot((np.eye(3)-np.dot(K,H)),temp_p)
    state=temp_x+np.dot(K,(z-temp_z))
    thet=thet+(0.01*state[2])
    
    vel.append(state)

vel=np.array(vel)
k=np.array(k)
print(k)
print(z)
```

```python
temp_vel=np.array(temp_vel)
plt.plot(range(n+1),vel[:,0])
plt.plot(range(n),temp_vel[:,0])
#plt.plot(range(n),acc_x)
```
